```python
import folium
import streamlit as st
from streamlit_folium import st_folium

# 1. Define Surat Center Coordinates
SURAT_LAT = 21.1702
SURAT_LON = 72.8311
API_KEY = "YOUR_TOMTOM_API_KEY"

def create_traffic_map():
    # 2. Initialize the Map
    m = folium.Map(location=[SURAT_LAT, SURAT_LON], zoom_start=13)

    # 3. Add TomTom Live Traffic Layer
    # This URL fetches 'tiles' which cover the whole area automatically
    traffic_tile_url = (
        f"https://api.tomtom.com/traffic/map/4/tile/flow/absolute/{{z}}/{{x}}/{{y}}.png"
        f"?key={API_KEY}&thickness=5"
    )

    folium.TileLayer(
        tiles=traffic_tile_url,
        attr='TomTom Traffic',
        name='Live Traffic Flow',
        overlay=True,
        control=True
    ).add_to(m)

    return m

# 4. Display in Streamlit
st.title("Surat Live Traffic Emission Map")
map_obj = create_traffic_map()
st_folium(map_obj, width=800, height=600)
```
